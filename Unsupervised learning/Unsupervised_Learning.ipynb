# Unsupervised Learning

## K-Means Clustering

```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# K-Means Clustering

X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=0.60, random_state=42)
```

```python
X
```

```python
df = pd.DataFrame(X, columns=['Feature-1', 'Feature-2'])
df.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
K_range = range(1, 11)
```

```python
for k in K_range:
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(X_scaled)
  inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertia, marker='o')
```

```python
kmeans_final = KMeans(n_clusters=3, random_state=42)
cluster_lables = kmeans_final.fit_predict(X_scaled)
```

```python
cluster_lables
```

```python
df['clusters'] = cluster_lables
```

```python
df.head(10)
```

```python
sns.scatterplot(x=df['Feature-1'], y=df['Feature-2'], hue=df['clusters'], palette='viridis')
```

# DB Scan

```python
from sklearn.datasets import make_moons
```

```python
X_moon, y_true = make_moons(n_samples=500, noise=0.05, random_state=42)
```

```python
from sklearn.cluster import DBSCAN
```

```python
df = pd.DataFrame(X_moon, columns=['Feature-1', 'Feature-2'])
df.head()
```

```python
X_scaled_moon = scaler.fit_transform(df)
```

```python
kmeans_final = KMeans(n_clusters=2, random_state=42)
cluster_lables = kmeans_final.fit_predict(X_scaled_moon)
```

```python
df['kmeans_clusters'] = cluster_lables
```

```python
sns.scatterplot(x=df['Feature-1'], y=df['Feature-2'], hue=df['kmeans_clusters'], palette='rainbow')
```

```python
# Now use DB Scan

dbscan = DBSCAN(eps=0.3, min_samples=5)
cluster_labels = dbscan.fit_predict(X_scaled_moon)
```

```python
df['dbscan_clusters'] = cluster_labels
```

```python
df.head()
```

```python
sns.scatterplot(x=df['Feature-1'], y=df['Feature-2'], hue=df['dbscan_clusters'], palette='viridis')
```
